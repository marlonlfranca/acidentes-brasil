```python
# Importando as bibliotecas necess√°rias e criando o dataframe:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.cm import ScalarMappable

data = pd.read_csv("AccidentsPortugues.csv")
```

```python
# Qual a BR com mais acidentes com v√≠timas feridas?

# Filtrar e contar acidentes com feridos
acidentes_feridos = data[data['feridos'] > 0]
acidentes_feridos_br = acidentes_feridos['br'].astype(str).str.split('.').str[0].value_counts().nlargest(10)

# C√°lculo dos percentuais
total = acidentes_feridos_br.sum()
porcentagens = (acidentes_feridos_br / total * 100).round(1)
# Normaliza√ß√£o para as cores
norm = plt.Normalize(acidentes_feridos_br.min(), acidentes_feridos_br.max())
cores = plt.cm.coolwarm(norm(acidentes_feridos_br))
# Plotagem
plt.figure(figsize=(12, 6))
bars = plt.bar(acidentes_feridos_br.index.astype(str), acidentes_feridos_br, color=cores)
for bar, valor, pct in zip(bars, acidentes_feridos_br, porcentagens):
    plt.text(bar.get_x() + bar.get_width()/2,
             1000,
             f'{valor}\n({pct}%)',
             ha='center',
             va='bottom',
             fontsize=9,
             color='white')
plt.title('BRs com mais v√≠timas feridas')
plt.xlabel('BR')
plt.ylabel('Acidentes com v√≠timas feridas')
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Qual UF tem mais acidentes com feridos?

# Filtrar e contar acidentes com feridos por UF
acidentes_feridos_uf = acidentes_feridos["uf"].value_counts().nlargest(10)
# C√°lculo dos percentuais
total = acidentes_feridos_uf.sum()
porcentagens = (acidentes_feridos_uf / total * 100).round(1)
# Normaliza√ß√£o para as cores
norm = plt.Normalize(acidentes_feridos_uf.min(), acidentes_feridos_uf.max())
cores = plt.cm.coolwarm(norm(acidentes_feridos_uf))
# Plotagem
plt.figure(figsize=(12, 6))
bars = plt.bar(acidentes_feridos_uf.index.astype(str), acidentes_feridos_uf, color=cores)
for bar, valor, pct in zip(bars, acidentes_feridos_uf, porcentagens):
    plt.text(bar.get_x() + bar.get_width()/2,
             500,
             f'{valor}\n({pct}%)',
             ha='center',
             va='bottom',
             fontsize=9,
             color='white')
plt.title('BRs com mais acidentes com v√≠timas feridas')
plt.xlabel('BR')
plt.ylabel('Acidentes com mais v√≠timas feridas')
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Total de acidentes com feridos por ano?
acidentes_feridos_acumulado_mes = pd.to_datetime(acidentes_feridos['data_inversa']).dt.month.value_counts().sort_values(ascending=False)

# C√°lculo dos percentuais
total = acidentes_feridos_acumulado_mes.sum()
porcentagens = (acidentes_feridos_acumulado_mes / total * 100).round(1)
# Normaliza√ß√£o para as cores
norm = plt.Normalize(acidentes_feridos_acumulado_mes.min(), acidentes_feridos_acumulado_mes.max())
cores = plt.cm.coolwarm(norm(acidentes_feridos_acumulado_mes))
# Plotagem
plt.figure(figsize=(12, 6))
bars = plt.bar(acidentes_feridos_acumulado_mes.index.astype(str), acidentes_feridos_acumulado_mes, color=cores)
for bar, valor, pct in zip(bars, acidentes_feridos_acumulado_mes, porcentagens):
    plt.text(bar.get_x() + bar.get_width()/2,
             1000,
             f'{valor}\n({pct}%)',
             ha='center',
             va='bottom',
             fontsize=9,
             color='black')
plt.title('Acidentes por M√™s (Acumulado)')
plt.xlabel('Acumulado por m√™s do ano (2017 - 2023)')
plt.ylabel('N√∫mero de Acidentes')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```
