### Com o ETL pronto, algumas perguntas surgem e a melhor forma de respondê-las é utilizando gráficos!

```python
# Importando as bibliotecas necessárias e criando o dataframe:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.cm import ScalarMappable

data = pd.read_csv("AccidentsPortugues.csv")
```

```python
data['data_inversa'] = pd.to_datetime(data['data_inversa'])

data_2022 = data[data['data_inversa'].dt.year == 2022].copy()
```

```python
# Qual a BR com mais acidentes?
acidentes_br = data_2022['br'].value_counts().nlargest(10)

# Normalização
norm = plt.Normalize(acidentes_br.min(), acidentes_br.max())
cores = plt.cm.coolwarm(norm(acidentes_br))  # Gradiente azul → vermelho

# Plotagem
plt.figure(figsize=(10, 6))
plt.bar(acidentes_br.index.astype(str), acidentes_br, color=cores)
plt.title('Acidentes por BR')
plt.xlabel('BR')
plt.ylabel('Número de Acidentes')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Quantos acidentes têm vítimas feridas, fatais ou não possui vítimas?
class_acidente = data_2022["classificacao_acidente"].value_counts().nlargest(10)

# Normalização
norm = plt.Normalize(class_acidente.min(), class_acidente.max())
cores = plt.cm.coolwarm(norm(class_acidente))  # Gradiente azul → vermelho

# Plotagem
plt.figure(figsize=(10, 6))
plt.bar(class_acidente.index.astype(str), class_acidente, color=cores)
plt.title('Acidentes por Classe')
plt.xlabel('Classe')
plt.ylabel('Número de Acidentes')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```
