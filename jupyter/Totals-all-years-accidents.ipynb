### Com o ETL pronto, algumas perguntas surgem e a melhor forma de respondê-las é utilizando gráficos!

```python
# Importando as bibliotecas necessárias e criando o dataframe:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.cm import ScalarMappable

data = pd.read_csv("AccidentsPortugues.csv")
```

```python
# Qual a BR com mais acidentes?
data['br'] = data['br'].astype(str).str.split('.').str[0]
acidentes_br = data['br'].value_counts().nlargest(10)
# Cálculo dos percentuais
total = acidentes_br.sum()
porcentagens = (acidentes_br / total * 100).round(1)
# Normalização para as cores
norm = plt.Normalize(acidentes_br.min(), acidentes_br.max())
cores = plt.cm.coolwarm(norm(acidentes_br))
# Plotagem
plt.figure(figsize=(10, 6))
bars = plt.bar(acidentes_br.index.astype(str), acidentes_br, color=cores)
for bar, valor, pct in zip(bars, acidentes_br, porcentagens):
    plt.text(bar.get_x() + bar.get_width()/2,
             2000,
             f'{valor}\n({pct}%)',
             ha='center',
             va='bottom',
             fontsize=9,
             color='white')
plt.title('Acidentes por BR')
plt.xlabel('BR')
plt.ylabel('Número de Acidentes')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Total de acidentes por ano?
data_backup2 = data.copy()
data['data_inversa'] = pd.to_datetime(data['data_inversa'])

acidentes_por_ano = data['data_inversa'].dt.year.value_counts().sort_index()

# Cálculo dos percentuais
total = acidentes_por_ano.sum()
porcentagens = (acidentes_por_ano / total * 100).round(1)
# Normalização para as cores
norm = plt.Normalize(acidentes_por_ano.min(), acidentes_por_ano.max())
cores = plt.cm.coolwarm(norm(acidentes_por_ano))
# Plotagem
plt.figure(figsize=(10, 6))
bars = plt.bar(acidentes_por_ano.index.astype(str), acidentes_por_ano, color=cores)
for bar, valor, pct in zip(bars, acidentes_por_ano, porcentagens):
    plt.text(bar.get_x() + bar.get_width()/2,
             2000,
             f'{valor}\n({pct}%)',
             ha='center',
             va='bottom',
             fontsize=9,
             color='black')
plt.title('Acidentes por Ano')
plt.xlabel('Ano')
plt.ylabel('Número de Acidentes')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Total de acidentes por mês (acumulado):
acidentes_por_mes_acumulado = data['data_inversa'].dt.month.value_counts().sort_values(ascending=False)

# Cálculo dos percentuais
total = acidentes_por_mes_acumulado.sum()
porcentagens = (acidentes_por_mes_acumulado / total * 100).round(1)
# Normalização para as cores
norm = plt.Normalize(acidentes_por_mes_acumulado.min(), acidentes_por_mes_acumulado.max())
cores = plt.cm.coolwarm(norm(acidentes_por_mes_acumulado))
# Plotagem
plt.figure(figsize=(10, 6))
bars = plt.bar(acidentes_por_mes_acumulado.index.astype(str), acidentes_por_mes_acumulado, color=cores)
for bar, valor, pct in zip(bars, acidentes_por_mes_acumulado, porcentagens):
    plt.text(bar.get_x() + bar.get_width()/2,
             2000,
             f'{valor}\n({pct}%)',
             ha='center',
             va='bottom',
             fontsize=9,
             color='black')
plt.title('Acidentes por Mês (Acumulado)')
plt.xlabel('Acumulado por mês do ano (2017 - 2023)')
plt.ylabel('Número de Acidentes')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```
